```python
#Data Preprocessing - change string values to numerical values
XData['Class/ASD']=XData['Class/ASD'].map({'YES':1,'NO':0})
XData['Gender']=XData['Gender'].map({'m':1,'f':0})
XData['Ethnicity']=XData['Ethnicity'].map({'Hispanic':0,'Black':1,'White-European':2,'\'Middle Eastern \'':3, '\'South Asian\'':4, 'Latino':0, 'Asian':4, 'Others':5})
XData['Jundice']=XData['Jundice'].map({'yes':1,'no':0})
XData['Family with PDD']=XData['Family with PDD'].map({'yes':1,'no':0})
#XData['Country of Res']=XData['Country of Res'].map({'Austria':0,'AmericanSamoa':1,'\'United Kingdom\'':2,'Albania':3, 'Belgium':4,'Afghanistan':5, 'Australia':6,'Bahrain':7, 'Azerbaijan':8, '\'United Arab Emirates\'':9, '\'New Zealand\'':10,'\'United States\'':11, 'Argentina':12,  })
XData['Used App Before']=XData['Used App Before'].map({'yes':1,'no':0})
XData['Relation']=XData['Relation'].map({'Parent':0,'Relative':1, 'Self':2, '\'Health care professional\'':3, 'Others':4})

XData
```

```python
#Revise No New NA Values where added after transforming the strinng values to nuemrical values
XData.isnull().sum()
```

```python
#Remove Missing Values
XClean = XData.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
XClean
```

```python
#Drop Duplicated Columns and Separate into X and Y components for data processing
Y = XClean['Class/ASD']
X = XClean.drop(columns=['Class/ASD','Age Des', 'Country of Res'])
X
```

```python
#Change values so that the previous transformation won't affect the training
col_names= XClean.select_dtypes("object").columns
onehot= pd.get_dummies(XClean, columns=col_names)
onehot
```

# AUTOENCODER

```python
#Split Data into Training and Testing (Change for experiments)
X_train, X_test, Y_train, Y_test = train_test_split(onehot,Y, test_size= 0.02, random_state=0)

def simpleAE(x,Y,encodersize, activationH, activationout):
    #create input layer
    input_layer = Input(shape=(x.shape[1],))
    #create encoder layer
    encoder_layer = Dense(encodersize, activation=activationH)(input_layer)
    #create decoder layer 
    decoder_layer = Dense(x.shape[1],activation = activationout)(encoder_layer)
    
    #create autoencoder and encoder models
    Autoencoder = Model(input_layer, decoder_layer)
    Encoder = Model(input_layer, encoder_layer)
    
    Autoencoder.compile(optimizer='adam', loss = 'mean_squared_error')
    
    
    Autoencoder.fit(X_train, X_train,
                   epochs = 50,
                   batch_size= 30,
                   shuffle=True,
                   verbose=1,
                   validation_data=(X_test, X_test))
    return Autoencoder, Encoder
```

```python
#Call Autoencoder Function
(AE, E) = simpleAE(onehot,Y,20,'relu','relu')
```

```python
#Save the Learned features to use in the NN
learnedfeatures = E.predict(X_train)
ValidationFeatures = E.predict(X_test)
print(learnedfeatures)
```

# Neural Network

```python
NN = MLPClassifier(solver='sgd',alpha=1e-6, hidden_layer_sizes=(28,1),learning_rate ='adaptive', random_state=1)
NN.fit(learnedfeatures,Y_train)
NNPredicted = NN.predict(ValidationFeatures)
print(NNPredicted, Y_test)
NNAccuracy = accuracy_score(Y_test,NNPredicted)*100
print(NNAccuracy)
```
